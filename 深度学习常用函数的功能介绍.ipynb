#### normal函数
> np.random.normal 函数用于从正态（高斯）分布中生成随机数。正态分布由两个参数确定：均值（mean）和标准差（standard deviation）。

> loc: float 或 array_like, 可选，默认值为 0.0。正态分布的均值（中心点）。
scale: float 或 array_like, 可选，默认值为 1.0。正态分布的标准差（分布的宽度，越大越扁平，越小越尖锐）。必须为非负数。
size: int 或 tuple of ints, 可选。输出的形状。如果给定 (m, n, k)，则生成一个形状为 (m, n, k) 的数组。默认情况下，将返回一个标量。


```python
# 生成一个标准正态分布的随机数
import numpy as np
random_number=np.random.normal()
print(random_number)
```

```python
# 生成形状为(3,3)的标准正态分布随机数组
random_matrix=np.random.normal(size=(3,3))
print(random_matrix)
```

```python
# 生成均值为5,标准差为2的单个随机数
random_number=np.random.normal(loc=5,scale=2)
print(random_number)
```

```python
# 生成均值为0,标准差为1的形状为(2,4)的随机数组
random_matrix=np.random.normal(loc=0,scale=1,size=(2,4))
print(random_matrix)
```

```python
#可视化生成随机数
import matplotlib.pyplot as plt

random_numbers=np.random.normal(loc=0,scale=1,size=1000)
# 绘制直方图
plt.hist(random_numbers, bins=30, edgecolor='k')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title('Histogram of Normally Distributed Random Numbers')
plt.show()
```

### zip函数
>zip 函数是 Python 的一个内置函数，用于将多个可迭代对象（如列表、元组、字符串等）打包成一个个元组，然后返回这些元组组成的一个迭代器。zip 函数的作用是将不同可迭代对象中的元素配对组合在一起。

```python
list1 = [1, 2, 3]
list2 = ['a', 'b', 'c']

zipped = zip(list1, list2)
print(list(zipped))
```

```python
#不同长度的列表
list1 = [1, 2, 3, 4]
list2 = ['a', 'b', 'c']

zipped = zip(list1, list2)
print(list(zipped))
```

```python
#多个列表
list1 = [1, 2, 3]
list2 = ['a', 'b', 'c']
list3 = [True, False, True]

zipped = zip(list1, list2, list3)
print(list(zipped))
```

### stack函数
>torch.stack 函数是 PyTorch 中的一个函数，用于沿新维度连接一系列张量。与 torch.cat 函数不同，torch.stack 会在指定维度上插入一个新的维度，然后在这个新的维度上连接张量。
>


```python
#沿新维度堆叠张量
import torch

# 创建两个形状相同的张量
tensor1 = torch.tensor([[1, 2], [3, 4]])
tensor2 = torch.tensor([[5, 6], [7, 8]])

# 将张量放入列表
tensors = [tensor1, tensor2]

# 沿第 0 维堆叠张量
stacked = torch.stack(tensors, dim=0)
print(stacked)
print(stacked.shape)
```

### enumrate函数
>是Python内置的一个函数,用于在遍历可迭代对象(如列表,元组或字符串)时生成一个索引序列和对应的元素,它返回一个迭代器,迭代时每次返回一个包含索引和元素的元组
>

```python
fruits=['apple','banana','cherry']
for idx,fruit in enumerate(fruits):
    print(idx,fruit)
```

```python
fruits = ['apple', 'banana', 'cherry']
for index, fruit in enumerate(fruits, start=1):
    print(index, fruit)
```

### DataLoader
>DataLoader的作用:
>* 批量加载数据:将数据集分成小批次,每次迭代返回一个批次的数据,这样可以有效利用GPU的并行计算能力
>* 打乱数据顺序:通过设置shuffle=True,可以在每个epoch开始时打乱数据顺序,防止模型过拟合顺序模式
>* 并行加载数据:通过设置num_workers可以利用多个子进程并行加载数据,提高数据加载效率
>* 自动批次处理:自动将数据分成指定大小的批次,并将数据转换为张量以供模型使用

```python
import torch
from torch.utils.data import DataLoader,TensorDataset

#数据和标签
data = torch.tensor([[1, 2], [3, 4], [5, 6], [7, 8]])
labels = torch.tensor([0, 1, 0, 1])

#创建一个TensorDataset对象
dataset=TensorDataset(data,labels)

#创建DataLoader对象用于按批次加载数据
train_iter=DataLoader(dataset,batch_size=2,shuffle=True)

#迭代Dataloader
#使用for循环遍历DataLoader,每次迭代返回一个批次的数据和标签
for i,(X,y) in enumerate(train_iter):
    print(f"Batch {i}:")
    print("x:",X)
    print("y:",y)
```
